## âš ï¸ Model Limitations

### 1. Historical Data Only
- **Issue**: Clustering based on past ~3 years
- **Impact**: Future risk may differ (market regime change)
- **Mitigation**: Regularly retrain model, monitor drift

### 2. No Fundamental Analysis
- **Issue**: Only uses price/volume technical data
- **Missing**: Earnings, debt, management quality, competitive moat
- **Impact**: Two stocks with similar technicals may have very different fundamentals
- **Mitigation**: Use clustering as initial filter, then do fundamental analysis

### 3. Market Context Ignored
- **Issue**: Doesn't account for overall market conditions
- **Example**: Low-risk stock in bull market might be high-risk in crash
- **Mitigation**: Add market regime detection, beta to market index

### 4. Limited Sample Size
- **Issue**: Only 57 NSE stocks (after filtering)
- **Impact**: May not capture full risk spectrum
- **Mitigation**: Include more stocks, use longer time periods

### 5. K-Means Assumptions
- **Issue**: Assumes spherical clusters, equal variance
- **Reality**: Risk profiles may have complex, non-spherical shapes
- **Alternative**: Try DBSCAN, Hierarchical clustering, Gaussian Mixture Models

### 6. Fixed Number of Clusters
- **Issue**: Pre-defined 4 clusters (subjective choice)
- **Impact**: Natural groupings might be 3 or 5 clusters
- **Mitigation**: Test multiple K values, use domain expertise
